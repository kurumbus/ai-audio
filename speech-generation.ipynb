```python
from transformers import pipeline
pipe = pipeline('text-to-speech')
```

```python
text = "sobaka, don't be sad, your uncle cat will help you with everything"
output = pipe(text)
output
```

```python
from IPython.display import Audio

Audio(data=output['audio'], rate=output['sampling_rate'])
```

```python
from pydub import AudioSegment
audio_seg = AudioSegment(output['audio'].tobytes(), frame_rate=output['sampling_rate'], sample_width=output['audio'].dtype.itemsize, channels=1)
audio_seg.export('suno_tts_example.mp3')
```
